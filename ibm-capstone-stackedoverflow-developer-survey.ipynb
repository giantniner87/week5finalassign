## Normalizing Data

<font size="3">There are two columns in the dataset that talk about compensation.</font>

 <font size="3">One is 'CompFreq'. This column shows how often a developer is paid (Yearly, Monthly, Weekly).</font>
 
 <font size="3">The other is 'CompTotal'. This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her 'CompFreq'.</font>
 
 <font size="3">This makes it difficult to compare the total compensation of the developers.</font>
 
 <font size="3">In this section we will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.</font>
 
 <font size="3">Once this column is ready, it makes comparison of salaries easy.</font>

```python
#Lets list out the various categories in the column 'CompFreq'
df['CompFreq'].unique()
```

```python
#Now lets create our new column named 'NormalizedAnnualCompensation'
df["CompFreq"].replace(to_replace="Yearly",value=1,inplace=True)
df["CompFreq"].replace(to_replace="Monthly",value=12,inplace=True)
df["CompFreq"].replace(to_replace="Weekly",value=52,inplace=True)
df['NormalizedAnnualCompensation'] = df["CompTotal"] * df["CompFreq"]
df.tail(5)
```

```python
#Display our new column 'NormalizedAnnualCompensation' median
df['NormalizedAnnualCompensation'].median()
```

```python
#Lets find out the Inter Quartile Range for the column 'ConvertedComp'
df['ConvertedComp'].describe()
```

```python
#Find out the upper and lower bounds
Q1 = df["ConvertedComp"].quantile(0.25)
Q3 = df["ConvertedComp"].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Identify how many outliers are there in the 'ConvertedComp' column
outliers = (df["ConvertedComp"] < (Q1 - 1.5 * IQR)) | (df["ConvertedComp"] > (Q3 + 1.5 * IQR))
outliers.value_counts()
```

```python
#Create a new dataframe by removing the outliers from the 'ConvertedComp' column
less = (df["ConvertedComp"] < (Q1 - 1.5 * IQR))
less.value_counts()

more = (df["ConvertedComp"] > (Q3 + 1.5 * IQR))
more.value_counts()

convertedcomp_out = df[~(df["ConvertedComp"] > (Q3 + 1.5 * IQR))]
convertedcomp_out.mean()
```

## Lets do some visualization through sqlite3!

```python
import sqlite3
conn=sqlite3.connect("../input/m4-survey-datasqlite/m4_survey_data.sqlite") #Open a database connection
```

```python
#Plot a histogram of 'ConvertedComp'
QUERY="""
SELECT *
FROM master
"""
df=pd.read_sql_query(QUERY,conn)
df.hist(figsize=(16,8), column='ConvertedComp', color='purple')

plt.title('ConvertedComp Histogram', fontsize=18)
plt.xlabel('ConvertedComp', fontsize=16)
plt.ylabel('Respondents', fontsize=16)
```

```python
#Create a box plot of 'Age'
QUERY="""
SELECT *
FROM master
"""

df=pd.read_sql_query(QUERY,conn)
df.boxplot(figsize=(16,8), column='Age', fontsize=14, color='green', vert=False)

plt.title('Age Box Plot', fontsize=18)
plt.xlabel('Age', fontsize=16)

```

```python
#Create a Scatterplot of 'Age' and 'WorkWeekHrs'
plt.figure(figsize=(16,8))

QUERY="""
SELECT *
FROM master
"""

df=pd.read_sql_query(QUERY,conn)
plot= sns.scatterplot(x='Age', y='WorkWeekHrs', data=df, alpha=0.50, color='darkblue')


plt.title('Age and WorkWeekHrs correlation', fontsize=18)
plt.ylabel('WorkWeekHrs', fontsize=16)
plt.xlabel('Age', fontsize=16)

plt.show()
```

```python
#Create a bubble plot of 'WorkWeekHrs' and 'CodeRevHrs', use 'Age' column as bubble size
    #First we have to normalize the column 'Age'
QUERY="""
SELECT WorkWeekhrs, CodeRevHrs, Age
FROM master
"""
df=pd.read_sql_query(QUERY, conn)
    
norm_age=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())

ax0=df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', figsize=(16,8), alpha=0.50, color='coral', fontsize=14, s=norm_age*1000+10, xlim=(0,200))
ax0.set_ylabel('CodeRevHrs', fontsize=16)
ax0.set_xlabel('WorkWeekHrs', fontsize=16)
ax0.set_title('Age bubbles on CodeRevHrs and WorkWeekHrs', fontsize=18)

plt.show()
```

```python
#Create a pie chart of top 5 databases desired to learn next year
table_name='DatabaseDesireNextYear'
QUERY="""
SELECT sql
FROM sqlite_master
WHERE name='{}'
""".format(table_name)

df=pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])

QUERY="""
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count(DatabaseDesireNextYear) DESC LIMIT 5
"""
df=pd.read_sql_query(QUERY,conn)
df.set_index('DatabaseDesireNextYear', inplace=True)

colors_list=['blue','yellow','green','orange','red','purple']


df['count'].plot(kind='pie', figsize=(18,6), autopct='%1.1f%%', startangle=90, shadow=True, labels=None, pctdistance=1.12, colors=colors_list, fontsize=14)

plt.title('Top 5 Databases Respondents Wish to Learn Next Year', y=1.12, fontsize=18)
plt.axis('equal')
plt.legend(labels=df.index, loc='upper left', fontsize=16)

plt.show()
```

```python
#Create a stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY="""
SELECT WorkWeekHrs, CodeRevHrs, AGE
FROM master
WHERE Age BETWEEN 30 and 35
"""
df=pd.read_sql_query(QUERY,conn)
df1=df.groupby('Age').median()

df1.plot(kind='bar', figsize=(16,8), stacked=True, fontsize=14)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35', fontsize=18)
plt.legend(loc='lower right', fontsize=16)
plt.xlabel('Age', fontsize=16)

plt.show()
```

```python
#Plot the median 'ConvertedComp' for all ages from 45 to 60
QUERY="""
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 and 60
"""
df=pd.read_sql_query(QUERY, conn)
df1=df.groupby('Age').median()

df1.plot(kind='line', figsize=(18,6), color='darkgreen', fontsize=14)

plt.title('Median ConvertedComp for Ages 45-60', fontsize=18)
plt.ylabel('ConvertedComp', fontsize=16)
plt.xlabel('Age', fontsize=16)

plt.show()
```

```python
#Plot a horizontal bar chart using column 'MainBranch'
QUERY="""
SELECT MainBranch, COUNT(*) as MainBranch
FROM master
GROUP BY MainBranch
"""
df=pd.read_sql_query(QUERY, conn)
df.plot(kind='barh', figsize=(10,6), color='lightblue', fontsize=14)

plt.title('Number of Respondents by Main Branch', fontsize=18)
plt.xlabel('Number of Respondents', fontsize=16)

plt.show()
```
